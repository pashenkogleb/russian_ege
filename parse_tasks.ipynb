```python
import json
import re
from bs4 import BeautifulSoup
```

```python
class Task:
    def __init__(self, hint, prompt, answers, correct_answer):
        self.hint = hint
        self.prompt = prompt
        self.answers = answers
        self.correct_answer = correct_answer
    def __repr__(self):
        answers = ""
        for i,answer in enumerate(self.answers):
            answers += f"{i+1}) {answer}\n"
        answers = answers.strip()
        #answers = "\n".join(self.answers)
        return f'''
{self.hint}
******************************************
{self.prompt}
******************************************
{answers}
******************************************
correct: {self.correct_answer}
'''
```

```python
def clean_answer(answer):
    answer = answer.strip()
    answer= re.sub("MathType@[^\s]+", "", answer)
    answer= re.sub("\xa0", "", answer)
    return answer

def parse_task(task):
    '''
    parses into task object
    '''
    parsed = BeautifulSoup(task['html'])
    #prompt_el = parsed.find("p",attrs = {"class":lambda x: x.lower()=="basis"})
    #prompt_el = parsed.find("p",attrs = {"class":lambda x: x.lower()=="basis"})
    prompt_el = parsed.find("p")

    hint = parsed.find("div", attrs = {"class":"hint"}).text
    assert prompt_el is not None
    prompt = clean_answer(prompt_el.text)
    answers = [clean_answer(x.text) for x in parsed.find_all("div", attrs = {"class":'answer'})]
    correct_answer = task['answer']
    return Task(hint, prompt, answers,correct_answer)
    


```

```python
file_name=  "data.json"

jsons =[]
with open(file_name, "r") as f:
    for line in f:
        jsons.append(json.loads(line))
```

```python
all_tasks = []
for obj in jsons:
    all_tasks +=(obj['tasks'])
```

```python
for i, task in enumerate(all_tasks):
    print(i)
    print(parse_task(task))
```
